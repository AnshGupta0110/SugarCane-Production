```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
df.head()
```

```python
df.shape
```

 # Data Cleaning

`Here we removed "." with "" (emplty string) also "," with "." in  Production (Tons), Production per Person (Kg), Acreage (Hectare), Yield (Kg / Hectare) of dataset Using ".srt.replace()" function`

```python
df['Production (Tons)'] = df['Production (Tons)'].str.replace(".","")
df['Production per Person (Kg)'] = df['Production per Person (Kg)'].str.replace(".","").str.replace(",",".")
df['Acreage (Hectare)'] = df['Acreage (Hectare)'].str.replace(".","")
df['Yield (Kg / Hectare)'] = df['Yield (Kg / Hectare)'].str.replace(".","").str.replace(",",".")

```

```python
df.head()
```

`Here we removed extra spaces(" ") from columns using "rename()" function `

```python
df.rename(columns = {"Production (Tons)":"Production(Tons)"}, inplace = True)
df.rename(columns = {"Production per Person (Kg)":"Production_per_person(Kg)"}, inplace = True)
df.rename(columns = {"Acreage (Hectare)":"Acreage(Hectare)"}, inplace = True)
df.rename(columns = {"Yield (Kg / Hectare)":"Yield(Kg/Hectare)"}, inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

`After checking null value we found two null values which has less impact on Data set so we remove these null values at 99th index country(Djibouti) using "dropna()" function and reset index.`

```python
df[df["Acreage(Hectare)"].isnull()]
```

```python
df = df.dropna().reset_index()
```

```python
df.head()
```

Here we can see that "index" and "Unnamed: 0" these are unnecessary so we gonna remove them Using "drop()" function,<br>
where,<br> we have two coulmns simultaneously so we use []<br>
 axie = 1 for coulmns<br> 


```python
df.drop(["index","Unnamed: 0"],axis = 1, inplace = True)
df.head()
```

```python
df.dtypes
```

`Converting string d-type to float For Numerical Analysis of Production(Tons), 
Production_per_person(Kg), Acreage(Hectare) and Acreage(Hectare) using "astype()" Function
`

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Production_per_person(Kg)"] = df["Production_per_person(Kg)"].astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(float)
df["Yield(Kg/Hectare)"] = df["Yield(Kg/Hectare)"].astype(float)
```
